```python
# Importing Pandas and Numpy libraries
import pandas as pd
import numpy as np
```

```python
# creating Series
sales1 = pd.Series([100, 200, 300, 400], index=['Microsoft', 'Facebook', 'Amazon', 'Google'])
sales2 = pd.Series([150, 240, 410, 490], index=['Microsoft', 'Facebook', 'Amazon', 'Google'])
sales3 = pd.Series([300, 400, 500, 600], index=['Microsoft', 'Facebook', 'Amazon', 'Google'])
```

```python
# Creating Dataframe
dataframe1 = pd.DataFrame({2017: sales1, 2018: sales2, 2019: sales3})
dataframe1
```

```python
# creating second Series
second_sales1 = pd.Series([150, 400, 200, 800], index=['Uber', 'Ola', 'Jugnoo', 'MyTaki'])
second_sales2 = pd.Series([150, 460, 410, 990], index=['Uber', 'Ola', 'Jugnoo', 'MyTaki'])
second_sales3 = pd.Series([300, 450, 500, 1000], index=['Uber', 'Ola', 'Jugnoo', 'MyTaki'])
```

```python
# Creating second Dataframe
dataframe2 = pd.DataFrame({2017:second_sales1, 2018: second_sales2, 2019: second_sales3})
dataframe2
```

```python
# Using append method
append1 = dataframe1.append(dataframe2)
append1
```

```python
append2 = dataframe2.append(dataframe1)
append2
```

```python
# append using different columns
dataframe2 = pd.DataFrame({2014:second_sales1, 2015: second_sales2, 2016: second_sales3})
dataframe2
```

```python
dataframe1
```

```python
append3 = dataframe1.append(dataframe2)
```

```python
append3
```

```python
dataframe1.append(dataframe2,ignore_index= True) # String Index is converted into numeric index staring from 0.
```

```python
dataframe1
```

```python
# example of loc()
dataframe1.loc['Microsoft',2017] = 200 
dataframe1
```

```python
# example of iloc()
fb_position = dataframe1.index.get_loc('Facebook')
year2017 = dataframe1.columns.get_loc(2017)
dataframe1.iloc[fb_position,year2017] = 350
dataframe1
```
