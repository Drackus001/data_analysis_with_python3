```python
# Importing pandas and numpy Libraries
import pandas as pd
import numpy as np
```

```python
Dataframe1 = pd.DataFrame( [[10, 20, 30], [40, 50, 60], [70, 80, 90]], index=[1, 2, 3],columns=['a', 'b','c'])
Dataframe1
```

```python
Dataframe2 = pd.DataFrame( [[100, 200, 300], [400, 500, 600], [700, 800, 900]], index=[1, 2, 3],columns=['a', 'b','c'])
Dataframe2
```

```python
pd.concat([Dataframe1, Dataframe2]) 
```

```python
pd.concat([Dataframe2, Dataframe1])
```

```python
Dataframe3 = pd.DataFrame( [[10, 20, 30], [40, 50, 60], [70, 80, 90]], index=[1, 2, 3],columns=['a', 'b','d'])
Dataframe3
```

```python
pd.concat([Dataframe2, Dataframe3], sort=False) # sort=False is optional
```
