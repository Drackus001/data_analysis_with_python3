```python
# Importing Libraries
import pandas as pd
import numpy as np
import datetime
from datetime import datetime
```

```python
# matplotlib
import matplotlib as mpl

# matplotlib plotting functions
import matplotlib.pyplot as plt

# we want our plots inline
%matplotlib inline
```

```python
Series1 = pd.Series(np.random.randn(911), index=pd.date_range("2017-1-1", "2019-6-30"))
Series2 = Series1.cumsum()
Series2.plot()
```

```python
Dataframe1 = pd.DataFrame(np.random.randn(911,2), index=Series1.index, columns =list('AB'))
Dataframe2 = Dataframe1.cumsum()
Dataframe2.plot()
```
