```python
# Importing pandas and numpy Libraries
import pandas as pd
import numpy as np
```

```python
Data = pd.read_csv("26_grouping_data.csv")
V1 = Data.copy() # for copy whole data
V1 = V1.set_index(['website','traffic']) # setting multiple index
V2 = V1.groupby(level=['website', 'traffic']) # multiple grouping (hierarchical grouping)
V2.count()
```

```python
V2.agg(np.sum) # aggregation of sum of two same group (ads= 30+40)
```

```python
V2.agg([np.mean,np.std, np.sum]) # multiple aggregation function.
```
