```python
Series1 = pd.Series(np.random.randn(911), index=pd.date_range("2017-1-1", "2019-6-30"))
Dataframe1 = pd.DataFrame(np.random.randn(911,2), index=Series1.index, columns =list('AB'))
Dataframe2 = Dataframe1.cumsum()
Dataframe2.plot()
```

```python
Dataframe2.plot(title="My_Title") # for adding title.
```
