```python
# Importing pandas and numpy Libraries
import pandas as pd
import numpy as np
```

```python
Data = pd.DataFrame({'Company':['Asus', 'HP', 'Apple', 'Dell', 'Lenovo'],
                    'Graphic_Card':['N', 'R', 'R', 'N', 'N'],
                    'Ci5': [1600,1700,np.nan,1900,2000],
                    'Ci7': [2100, 2200, 2300, 2400, 2500]})
Data
```

```python
V1 = Data.groupby('Company')
V1.count()
```

```python
V1.transform(lambda x:x+10)
```

```python
V2 = V1.transform(lambda x:x.fillna(x.mean())) # mean of only one NaN group which is also NaN
V2
```
